# âš™ï¸ Activation Functions in ANN

Activation functions introduce **non-linearity** into neural networks, allowing them to learn complex patterns. Without them, even deep networks behave like linear models.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## ðŸ” Why Activation Functions?

They help the network learn from **non-linear data** by transforming the input signal before passing it to the next layer.

Without activation:
\[
f(x) = Wx + b \Rightarrow \text{Linear only}
\]

With activation:
\[
f(x) = \phi(Wx + b) \Rightarrow \text{Can model non-linear boundaries}
\]

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

## ðŸ“Š Visualizing Activation Functions

Let's compare the shapes of commonly used activation functions.

```python
x = np.linspace(-10, 10, 1000)

plt.figure(figsize=(10, 6))
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, leaky_relu(x), label='Leaky ReLU')
plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

## âœ… Summary of Common Activation Functions

| Function     | Range        | Non-linearity | Use-case                        |
|--------------|--------------|----------------|----------------------------------|
| Sigmoid      | (0, 1)       | Yes            | Binary classification output     |
| Tanh         | (-1, 1)      | Yes            | Hidden layers (early usage)      |
| ReLU         | [0, âˆž)       | Yes            | Most hidden layers (default)     |
| Leaky ReLU   | (-âˆž, âˆž)      | Yes            | Prevents dying ReLU problem      |

```python
input_data = np.array([-2, -1, 0, 1, 2])
output = relu(input_data)
print(f"ReLU({input_data}) = {output}")
```
